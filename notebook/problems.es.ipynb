## Ejercicio 1 

Dos dados se lanzan una vez y se observa el total obtenido. Usa una simulación para encontrar la probabilidad estimada de que la puntuación sea mayor a 7 o un número par. Una simulación es una repetición del mismo experimento múltiples veces para observar su comportamiento:

- Ejecuta el experimento 1000 veces (lanza 2 dados 1000 veces, y suma el número de ambos dados).
- Lleva la cuenta de los números y las veces que la suma fue mayor a 7 o un número par.
- Divide el número del paso 2 entre el número de iteraciones(1000).

```python
# TODO
import numpy as np
dado_1 = range(1, 7)
dado_2 = range(1, 7)
tirada_dado1 = np.random.choice(dado_1, size=1000)
tirada_dado2 = np.random.choice(dado_2, size=1000)

#print(tirada_dado1)
sum_tirada_dado1 = np.sum(dado_1)
sum_tirada_dado2 = np.sum(dado_2)
tirada_dados_total = tirada_dado1 + tirada_dado2

#print(sum_tirada_dado1)

#tirada_dados_total me da que no xD

tirada_mas_7 = np.sum(tirada_dados_total > 7)
tirada_par = np.sum(tirada_dados_total %2 == 0)
tirada_mas_7_final = (tirada_mas_7 / 1000)*100
tirada_par_final = (tirada_par / 1000)*100
par_y_mas_7 = np.sum((tirada_dados_total > 7) & (tirada_dados_total % 2 == 0))
par_y_mas_final = (par_y_mas_7 / 1000)*100

print(f"La probabilidad de que salgan mayores de 7 será de {tirada_mas_7_final:.2f} %")
print(f"La probabilidad de que salgan pares será de {tirada_par_final:.2f} %")
print(f"La probabilidad de que en 1000 tiradas los dados sean pares y mayores que 7 es de {par_y_mas_final:.2f} %")






```

```python
import random

tiradas = 1000

mayor_7 = 0
pares = 0
pares_y_mayor_7 = 0

for _ in range(tiradas):
    dado_1 = random.randint(1,7)
    dado_2 = random.randint(1,7)
    suma = dado_1 + dado_2
    if suma > 7:
        mayor_7 += 1
    if suma %2 == 0:
        pares +=1
    if suma > 7 and suma %2 == 0:
        pares_y_mayor_7 += 1

#print(f"La suma fue mayor que 7 en {mayor_7} tiradas.")
#print(f"La suma fue un número par en {pares} tiradas.")
#print(f"La suma fue un número par y mayor que 7 en {pares_y_mayor_7} tiradas.")    

mayor_7_prob = (mayor_7 / 1000) *100
pares_prob = (pares / 1000) *100
pares_y_mayor_7_prob = (pares_y_mayor_7 / 1000) *100

print(f"La probabilidad de que sea mayor que 7 es de  {mayor_7_prob:.2f} %.")
print(f"La probabilidad de que sea par es de {pares_prob} %.")
print(f"La probabilidad de que sea número par y además mayor que 7 es {pares_y_mayor_7_prob:.2f} %.")    


```

## Ejercicio 2

Una caja contiene 10 bolas blancas, 20 bolas rojas y 30 bolas verdes. Si tomamos 5 bolas de la caja reemplazándolas (tomando una bola, anotando el color y luego regresandola a la caja). Queremos saber la probabilidad de:

1. Tomar 3 blancas y 2 rojas.
2. Todas del mismo color.

Ejecuta el experimento 1000 veces y calcula las probabilidades mencionadas.

```python
import random

# Crear la caja con bolas usando un diccionario
ball_box = {}

for i in range(60):
    if i < 10:
        ball_box[i] = "White"
    elif (i >= 10) and (i < 30):
        ball_box[i] = "Red"
    else:
        ball_box[i] = "Green"

# Convertimos el diccionario a una lista de bolas
caja_bolas = list(ball_box.values())

# Contadores para las dos condiciones
blancas_3_rojas_2 = 0
bolas_mismo_color = 0


for i in range(1000):
    # Seleccionar 5 bolas con reemplazo usando random.choices()
    bola_seleccionada = random.choices(caja_bolas, k=5)

    
    if bola_seleccionada.count("White") == 3 and bola_seleccionada.count("Red") == 2:
        blancas_3_rojas_2 += 1

    
    if len(set(bola_seleccionada)) == 1:
        bolas_mismo_color += 1


prob_blancas_3_rojas_2 = (blancas_3_rojas_2 / 1000) * 100
prob_bolas_mismo_color = (bolas_mismo_color / 1000) * 100


print(f"La probabilidad de que salgan 3 bolas blancas y 2 rojas es del {prob_blancas_3_rojas_2:.2f} %.")
print(f"La probabilidad de que salgan todas las bolas del mismo color es del {prob_bolas_mismo_color:.2f} %.")

```
