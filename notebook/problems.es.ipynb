## Ejercicio 1 

Dos dados se lanzan una vez y se observa el total obtenido. Usa una simulación para encontrar la probabilidad estimada de que la puntuación sea mayor a 7 o un número par. Una simulación es una repetición del mismo experimento multiples veces para observar su comportamiento:

- Ejecuta el experimento 1000 veces (lanza 2 dados 1000 veces, y suma el número de ambos dados).
- Lleva ua cuenta de los números y las veces que la suma fue mayor a 7 o un número par.
- Divide el número del paso 2 entre el número de iteraciones(1000).

```python
# TODO
import numpy as np
dado_1 = range(1, 7)
dado_2 = range(1, 7)
tirada_dado1 = np.random.choice(dado_1, size=1000)
tirada_dado2 = np.random.choice(dado_2, size=1000)

#print(tirada_dado1)
sum_tirada_dado1 = np.sum(dado_1)
sum_tirada_dado2 = np.sum(dado_2)
tirada_dados_total = tirada_dado1 + tirada_dado2

#print(sum_tirada_dado1)

#tirada_dados_total me da que no xD

tirada_mas_7 = np.sum(tirada_dados_total > 7)
tirada_par = np.sum(tirada_dados_total %2 == 0)
tirada_mas_7_final = (tirada_mas_7 / 1000)*100
tirada_par_final = (tirada_par / 1000)*100
par_y_mas_7 = np.sum((tirada_dados_total > 7) & (tirada_dados_total % 2 == 0))
par_y_mas_final = (par_y_mas_7 / 1000)*100

print(f"La probabilidad de que salgan mayores de 7 será de {tirada_mas_7_final} %")
print(f"La probabilidad de que salgan pares será de {tirada_par_final} %")
print(f"La probabilidad de que en 1000 tiradas los dados sean pares y mayores que 7 es de {par_y_mas_final} %")






```

```python
import random

tiradas = 1000

mayor_7 = 0
pares = 0
pares_y_mayor_7 = 0

for _ in range(tiradas):
    dado_1 = random.randint(1,7)
    dado_2 = random.randint(1,7)
    suma = dado_1 + dado_2
    if suma > 7:
        mayor_7 += 1
    if suma %2 == 0:
        pares +=1
    if suma > 7 and suma %2 == 0:
        pares_y_mayor_7 += 1

#print(f"La suma fue mayor que 7 en {mayor_7} tiradas.")
#print(f"La suma fue un número par en {pares} tiradas.")
#print(f"La suma fue un número par y mayor que 7 en {pares_y_mayor_7} tiradas.")    

mayor_7_prob = (mayor_7 / 1000) *100
pares_prob = (pares / 1000) *100
pares_y_mayor_7_prob = (pares_y_mayor_7 / 1000) *100

print(f"La probabilidad de que sea mayor que 7 es de  {mayor_7_prob} %.")
print(f"La probabilidad de que sea par es de {pares_prob} %.")
print(f"La probabilidad de que sea número par y además mayor que 7 es {pares_y_mayor_7_prob} %.")    


```
